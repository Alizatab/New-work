```python
# Python code used to generate Figure 1 (for context)
import matplotlib.pyplot as plt
import pandas as pd

# Data for the chart
methods = ['Case Studies', 'Field Visits', 'Hands-on Specimens', 'Inquiry-based']
gain_scores = [0.64, 0.59, 0.78, 0.76] # Approximate mean gains calculated from the dataset

plt.figure(figsize=(8, 5))
bars = plt.bar(methods, gain_scores, color=['skyblue', 'lightgreen', 'gold', 'lightcoral'])
plt.ylabel('Mean Cognitive Gain (Hedges g)')
plt.title('Figure 1: Cognitive Gains by Primary Instructional Method')
plt.ylim(0, 0.9)
for bar in bars:
    height = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2., height + 0.01,
             f'{height:.2f}', ha='center', va='bottom')
plt.tight_layout()
plt.show()
```
